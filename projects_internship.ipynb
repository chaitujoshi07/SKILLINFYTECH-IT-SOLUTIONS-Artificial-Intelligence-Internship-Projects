```python
##Title and information
#PROJECT 1: AI CHATBOT FOR COLLEGE HELPDESK
# Internship Project on artificial intelligence lerning Natural Language Understanding (NLU)
#Intent classification with MLRule-based AI and fallback handling Dialogue management

#1.setup environment
## Install required libraries
!pip install nltk scikit-learn tensorflow


#2.Import Libraries
import nltk
import numpy as np
import string
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity


#3.Prepare College Helpdesk FAQ Dataset
corpus = [
    "What are the college timings?",
    "The college is open from 9 AM to 5 PM.",
    "How can I apply for admission?",
    "You can apply online through the admission portal.",
    "What courses are offered?",
    "We offer B.Tech, MBA, and MCA programs.",
    "Where is the library located?",
    "The library is on the second floor of the main building.",
    "Who is the principal of the college?",
    "Dr. Sharma is the principal of the college."
]

#4.Preprocess Text
nltk.download('punkt')

def preprocess(text):
    return text.lower().translate(str.maketrans('', '', string.punctuation))

sent_tokens = [preprocess(sentence) for sentence in corpus]

#5.Build Chatbot Logic
def chatbot_response(user_input):
    user_input = preprocess(user_input)
    sent_tokens.append(user_input)

    vectorizer = TfidfVectorizer()
    tfidf = vectorizer.fit_transform(sent_tokens)

    vals = cosine_similarity(tfidf[-1], tfidf[:-1])
    idx = vals.argsort()[0][-1]
    flat = vals.flatten()
    flat.sort()
    score = flat[-1]

    sent_tokens.pop()  # remove user input from corpus

    if score == 0:
        return "I'm sorry, I don't understand. Please contact the helpdesk."
    else:
        return corpus[idx+1]  # return the answer


#6.Run Chatbot
def chat_with_bot():
    queries = [
        "What are the college timings?",
        "Where is the library?",
        "Who is the principal?",
        "bye"
    ]
    for q in queries:
        print("You:", q)
        if q.lower() == 'bye':
            print("Chatbot: Goodbye! Have a great day.")
            break
        else:
            print("Chatbot:", chatbot_response(q))

chat_with_bot()


#Expected Output

#You: What are the college timings?

#You: Where is the library?

#You: Who is the principal?

#You: bye

```

```python
#PROJECT 2:  Tic-Tac-Toe Game with AI Player
# Internship Project on artificial intelligence lerning Game theory
#Recursive algorithms (Minimax) AI move prediction Basic UI integration with adding GUI
# full game with scorecard and Difficulty Selector

#1: Setup Environment
!pip install ipywidgets

from IPython.display import display
import ipywidgets as widgets
import numpy as np


#2: Define Game Logic (Minimax + Random AI)
def create_board():
    return np.array([[' ']*3 for _ in range(3)])

def check_winner(board):
    for i in range(3):
        if board[i][0] == board[i][1] == board[i][2] != ' ':
            return board[i][0]
        if board[0][i] == board[1][i] == board[2][i] != ' ':
            return board[0][i]
    if board[0][0] == board[1][1] == board[2][2] != ' ':
        return board[0][0]
    if board[0][2] == board[1][1] == board[2][0] != ' ':
        return board[0][2]
    return None

def is_full(board):
    return not any(' ' in row for row in board)

def minimax(board, depth, is_maximizing):
    winner = check_winner(board)
    if winner == 'O': return 1
    elif winner == 'X': return -1
    elif is_full(board): return 0

    if is_maximizing:
        best_score = -np.inf
        for i in range(3):
            for j in range(3):
                if board[i][j] == ' ':
                    board[i][j] = 'O'
                    score = minimax(board, depth+1, False)
                    board[i][j] = ' '
                    best_score = max(score, best_score)
        return best_score
    else:
        best_score = np.inf
        for i in range(3):
            for j in range(3):
                if board[i][j] == ' ':
                    board[i][j] = 'X'
                    score = minimax(board, depth+1, True)
                    board[i][j] = ' '
                    best_score = min(score, best_score)
        return best_score

def ai_move(board, difficulty="hard"):
    if difficulty == "easy":
        # Random move
        empty = [(i,j) for i in range(3) for j in range(3) if board[i][j]==' ']
        if empty:
            move = empty[np.random.randint(len(empty))]
            board[move[0]][move[1]] = 'O'
    else:
        # Minimax (unbeatable)
        best_score = -np.inf
        move = None
        for i in range(3):
            for j in range(3):
                if board[i][j] == ' ':
                    board[i][j] = 'O'
                    score = minimax(board, 0, False)
                    board[i][j] = ' '
                    if score > best_score:
                        best_score = score
                        move = (i, j)
        if move:
            board[move[0]][move[1]] = 'O'


#3: Build GUI with ipywidgets with scorebord and Difficulty Selector
board = create_board()
buttons = [[widgets.Button(description=' ', layout=widgets.Layout(width='60px', height='60px')) for j in range(3)] for i in range(3)]
output = widgets.Output()

# Scoreboard
scores = {"X":0, "O":0, "Draw":0}
score_label = widgets.Label(value=f"Scores → Human (X): {scores['X']} | AI (O): {scores['O']} | Draws: {scores['Draw']}")

# Difficulty selector
difficulty_dropdown = widgets.Dropdown(options=["easy","hard"], value="hard", description="AI Difficulty:")

def update_score(winner):
    if winner == 'X':
        scores['X'] += 1
    elif winner == 'O':
        scores['O'] += 1
    else:
        scores['Draw'] += 1
    score_label.value = f"Scores → Human (X): {scores['X']} | AI (O): {scores['O']} | Draws: {scores['Draw']}"

def on_click(i, j):
    def handler(b):
        global board
        if board[i][j] == ' ' and not check_winner(board):
            board[i][j] = 'X'
            buttons[i][j].description = 'X'
            winner = check_winner(board)
            if winner or is_full(board):
                update_score(winner if winner else "Draw")
                with output: print("Winner:", winner if winner else "Draw")
                return
            ai_move(board, difficulty=difficulty_dropdown.value)
            for x in range(3):
                for y in range(3):
                    buttons[x][y].description = board[x][y]
            winner = check_winner(board)
            if winner or is_full(board):
                update_score(winner if winner else "Draw")
                with output: print("Winner:", winner if winner else "Draw")
    return handler

for i in range(3):
    for j in range(3):
        buttons[i][j].on_click(on_click(i,j))

grid = widgets.GridBox(children=[buttons[i][j] for i in range(3) for j in range(3)],
                       layout=widgets.Layout(grid_template_columns="60px 60px 60px"))


#expected outcome
#How It Works:
#A 3x3 grid of buttons appears in Colab.
#Click a cell to place your X.
#The AI responds with O using either:
#Easy mode → random moves.
#Hard mode → Minimax (unbeatable).
#The scoreboard updates after each round.
```

```python
#PROJECT 6: SMART TRAFFIC LIGHT CONTROLLER
#Internship Project on artificial intelligence lerning AI rule engine / Reinforcement
#logiC OpenCV object detection Simulation-based input/output Real-time data decision making Multi‑Objective AI.

#1.Setup Environment
!pip install numpy gym requests


#2. 2: Define Traffic Environment (Gym‑style)
#We’ll simulate a 4‑way intersection with vehicles arriving randomly.
import numpy as np
import gym
from gym import spaces

class TrafficEnv(gym.Env):
    def __init__(self):
        super(TrafficEnv, self).__init__()
        # Actions: 0=North-South green, 1=East-West green
        self.action_space = spaces.Discrete(2)
        # State: vehicle counts [N, S, E, W]
        self.observation_space = spaces.Box(low=0, high=50, shape=(4,), dtype=np.int32)
        self.reset()

    def reset(self):
        self.state = np.random.randint(0, 10, size=(4,))
        return self.state

    def step(self, action):
        # Vehicles arrive randomly
        arrivals = np.random.randint(0, 5, size=(4,))
        self.state += arrivals

        # Apply action: clear vehicles in chosen direction
        if action == 0:  # North-South green
            self.state[0] = max(0, self.state[0]-np.random.randint(3,7))
            self.state[1] = max(0, self.state[1]-np.random.randint(3,7))
        else:  # East-West green
            self.state[2] = max(0, self.state[2]-np.random.randint(3,7))
            self.state[3] = max(0, self.state[3]-np.random.randint(3,7))

        # Reward: negative of total waiting vehicles
        reward = -np.sum(self.state)

        done = False
        return self.state, reward, done, {}


#3.Reinforcement Learning Agent (Q‑Learning)
class QLearningAgent:
    def __init__(self, state_size, action_size):
        self.q_table = {}
        self.state_size = state_size
        self.action_size = action_size
        self.alpha = 0.1   # learning rate
        self.gamma = 0.9   # discount factor
        self.epsilon = 0.2 # exploration rate

    def get_state_key(self, state):
        return tuple(state)

    def choose_action(self, state):
        if np.random.rand() < self.epsilon:
            return np.random.randint(self.action_size)
        key = self.get_state_key(state)
        if key not in self.q_table:
            self.q_table[key] = np.zeros(self.action_size)
        return np.argmax(self.q_table[key])

    def learn(self, state, action, reward, next_state):
        key = self.get_state_key(state)
        next_key = self.get_state_key(next_state)
        if key not in self.q_table:
            self.q_table[key] = np.zeros(self.action_size)
        if next_key not in self.q_table:
            self.q_table[next_key] = np.zeros(self.action_size)
        predict = self.q_table[key][action]
        target = reward + self.gamma * np.max(self.q_table[next_key])
        self.q_table[key][action] += self.alpha * (target - predict)


#4.IoT Integration (Simulated Real API)
# HERE Traffic API or Google Maps Traffic API. Here’s a simulated IoT sensor function:
import requests

def iot_sensor_data():
    # Example: Replace with real API call
    # response = requests.get("YOUR_TRAFFIC_API_URL").json()
    # Extract traffic density from response
    return np.random.randint(0, 20, size=(4,))


#5.Multi‑Objective Reward Function
#Balance vehicles, pedestrians, and emergency vehicles:
def multi_objective_reward(state, pedestrians_waiting, emergency_priority):
    vehicle_penalty = -np.sum(state)          # fewer cars waiting
    pedestrian_penalty = -pedestrians_waiting # fewer pedestrians waiting
    emergency_bonus = emergency_priority * 10 # prioritize ambulances
    return vehicle_penalty + pedestrian_penalty + emergency_bonus


#6.Train RL Agent with IoT + Multi‑Objective AI
env = TrafficEnv()
agent = QLearningAgent(state_size=4, action_size=2)

episodes = 10
for ep in range(episodes):
    state = iot_sensor_data()  # IoT input
    total_reward = 0
    for t in range(20):
        action = agent.choose_action(state)
        next_state, _, _, _ = env.step(action)
        reward = multi_objective_reward(next_state, pedestrians_waiting=5, emergency_priority=1)
        agent.learn(state, action, reward, next_state)
        state = next_state
        total_reward += reward
    print(f"Episode {ep+1}, Total Reward: {total_reward}")


#7.Expected Behavior
#RL agent learns to switch lights optimally to reduce waiting vehicles.
#IoT sensors provide dynamic traffic data.
#Multi‑objective AI balances cars, pedestrians, and emergency vehicles.
```
